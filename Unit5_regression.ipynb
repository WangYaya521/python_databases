```python
# tips = sns.load_dataset("tips")  sns.load_dataset()是提取sns内部数据的

tips = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Seaborn\set1.csv")

tips.head()
```

```python
# 分析total_bill与tip之间的关系，回归分析(线性回归)
# regplot()

sns.set_style("ticks")

fig = plt.figure(figsize = (8 , 6))


sns.regplot(x = "total_bill" , y = "tip" , data = tips , color = "Red")

sns.despine()

sns.set_context("paper" , font_scale = 1.2 , rc = {"lines_linewidth" : 2})
```

```python
# sns.regplot(x = "size" , y = "tip" , data = tips)

sns.regplot(x = "size" , y = "tip" , data = tips , x_jitter = 0.1) # 在x上加上一个小浮动
```

```python
# data	必需，Pandas DataFrame 数据	-
# hue	按分类变量分组着色（最常用）	列名，如 'species'
# vars	仅绘制指定的数值列	列表，如 ['sepal_length', 'petal_width']
# kind	非对角线图类型	'scatter'(默认), 'reg'(带回归线), 'kde'(核密度图), 'hist'（直方图） ,  
# diag_kind	对角线图类型	'auto', 'hist', 'kde'
# palette	配色方案	'husl', 'viridis', 'Set2'
# height / aspect	子图高度 / 宽高比	数字，如 height=2
# corner	只画左下角三角（节省空间）	True / False
# plot_kws	自定义非对角线样式	{'alpha': 0.5, 's': 30} (透明度 / 点大小)

sns.pairplot(data = tips , kind = "kde")
```

```python
# 多变量分析

# 类别分析stripplot() , swarmplot()

day = tips["day"]
print(day.value_counts())

fig = plt.figure(figsize = (12 , 6))
ax1 = fig.add_subplot(1 , 2 , 1)
ax2 = fig.add_subplot(1 , 2 , 2)


sns.stripplot(data = tips , x = "day" , y = "total_bill" , palette = sns.color_palette("bright") , ax = ax1)#点图
ax1.set_title("strip")
ax1.tick_params(axis = "x" , rotation = 30)

sns.swarmplot(data = tips , x = "day" , y = "total_bill" , palette = sns.color_palette("Reds") , ax = ax2)#点树状

```

```python
# 额外加指标hue = ""

sns.swarmplot(data = tips , x = "day" , y = "total_bill" , palette = sns.color_palette("hls") , hue = "size")

plt.legend(loc = "upper right")
```

```python
# 盒图sns.boxplot

sns.boxplot(data = tips , x = "day" , y = "total_bill" , hue = "smoker" , palette = sns.color_palette("hls"))
```

```python
# 小提琴图，越胖数据越多

#sns.violinplot(data = tips , x = "day" , y = "total_bill" , hue = "smoker" , palette = sns.color_palette("hls"))

# 指定两边属性 ， split
sns.violinplot(data = tips , x = "day" , y = "total_bill" , hue = "smoker" , palette = sns.color_palette("hls") , split=True)
```

```python
# 两图合并

sns.swarmplot(data = tips , x = "day" , y = "total_bill"  , color = "w" , alpha = .5) # alpha透明程度

sns.violinplot(data = tips , x = "day" , y = "total_bill" , palette = sns.color_palette("hls"))
```

```python
# 条形图barplot

sns.barplot(data = tips , x = "sex" , y = "total_bill" , hue = "day" , palette = sns.color_palette("hls"))
```

```python
# 描述差异变化

sns.pointplot(data = tips , x = "day" , y = "total_bill" , hue = "smoker" , 
              palette = {"Yes" : "g" , "No" : "r"} ,  # 指定线的颜色
              markers = ["*" , "^"] , # 指定特征的形状
              linestyles = ["-" , "--"]); # 指定线形状
```
