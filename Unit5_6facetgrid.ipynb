```python
# 展示数据集中很多子集时，使用

import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Seaborn\set1.csv")

tips[2:15:2]

tips["smoker"][1:4]
```

```python
# 先指定需要的指标，给出空白图，再指定xy轴

g = sns.FacetGrid(data = tips , col = "sex" , hue = "smoker" , palette = sns.color_palette("hls"))
```

```python


# matplotlib/中的画图类型有：scatter（散点图） , bar(柱状图)，hist(单变量频率直方图) ， 
# kind are 'strip'(点图), 'swarm'(树状点图), 'box'(盒图), 'boxen', 'violin'(小提琴图)(split = True), 'bar'(柱状图), 'count'(单变量分析,统计数量), and 'point' , reg

# 先指定需要的指标，再指定绘图类型以及xy轴

g = sns.FacetGrid(data = tips , row = "sex" , hue = "smoker" , palette = sns.color_palette("hls")) # col = "" , row = ""都可以

#g.map(plt.scatter, "size" , "total_bill") # 这里底层是matplotlib,所以没有palette，应该在sns里面调用 ， 在g.map里面只指定类型，xy轴
g.map(sns.pointplot,"day" , "total_bill" )


g.add_legend()
plt.show()
```

```python
from pandas import Categorical

# 给出row以及顺序
#order_day = Categorical(["Thu" , "Fri" , "Sat" , "Sun"])
order_smoker = Categorical(["Yes" , "No"])

g = sns.FacetGrid(data = tips , col = "sex" , row = "smoker" , row_order = order_smoker, palette = sns.color_palette("hls"))

g.map(sns.violinplot, "day" , "total_bill" )

# 标注横纵坐标名称分别是什么 ， 如果不加以说明，默认g.map(sns.swarmplot, "size" , "total_bill" )中的size与total_bill
g.set_axis_labels("Day" , "total_bill(us)")

# 设置子图与子图之间的间隔
# g.fig.subplots_adjust(left = .5 , right = 1 , bottom = .5 , top = .9 , wspace=.2 , hspace=.2)
# 左右上下偏移（一般默认即可），子图与子图之间横纵间隔
g.fig.subplots_adjust( wspace=.2 , hspace=.2)

g.add_legend()
plt.show()
```

```python
f = sns.pairplot(data = tips , kind = "reg")

plt.show()
```

```python
#iris = sns.load_dataset("iris")

# 先指定绘图类型，pair所有变量之间的关系
f = sns.PairGrid(data = tips ,  hue = "smoker" , palette = sns.color_palette("hls"))

#再指定绘制类型，散点/回归等
# f.map(plt.scatter)
# f.map(sns.regplot)

#规定对角线什么类型，非对角线类型

#matplotlib
f.map_diag(plt.hist)
f.map_offdiag(plt.scatter)

#seaborn
# f.map_diag(sns.barplot)
# f.map_offdiag(sns.stripplot)

f.add_legend()
plt.show()
```

```python
# f = sns.PairGrid(data = tips , vars = ["total_bill" , "tips"](选取想要比较的特征) , hue = "smoker")

f = sns.PairGrid(data = tips , vars = ["total_bill" , "tip"],  hue = "smoker")

#指定绘制类型，散点/回归等
# f.map(plt.scatter)
f.map(sns.regplot)



f.add_legend()
plt.show()
```
