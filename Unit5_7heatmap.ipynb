```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.random.rand(8,8)

print(data)
```

```python
# 通过颜色差异表现出矩阵中数字的大小

heatmap = sns.heatmap(data)
```

```python
# 规定取值上下限vmin , vmax , 超出最大值颜色就是最大值对应 的颜色，低于最小值，就是最小值颜色

heatmap = sns.heatmap(data , vmin = 0.3 , vmax = 0.7)
```

```python
# 矩阵中有正有负

data = np.random.randn(8,8)

print(data)
```

```python
heatmap = sns.heatmap(data , center = 0 , vmin = -1.5 , vmax = 1.5)
```

```python
passages = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Seaborn\passengers.csv")

passages.head()
```

```python
# df.pivot(index=行, columns=列, values=值)
# pd.pivot_table(df, index=行, columns=列, values=值, aggfunc='mean/sum/count')

# 数据透视表.pivot_table(value = "需要分析的数值，销售额/薪资/需求量" , index = "按什么分组，比如岗位/性别/城市/" （行分组） ， 
#                               columns = "按特征分组"（列分组），  aggfunc = "对数据怎么处理，求和/平均值/归一")


passengers_matrix = passages.pivot(index = "month" , columns = "year" ,values = "passengers")

print(passengers_matrix)
```

```python
passengers_heatmap = sns.heatmap(passengers_matrix)

```

```python
# annot = True , 控制是否在每个单元格内显示具体数值。设为 True 表示显示。
# fmt="d"	格式控制	指定标注数值的字符串格式。
# - d 代表 decimal (整数) 强制将数值以整数形式显示，不显示小数位。
# annot_kws={"size": 10}：调整标注数字的字体大小

ax = sns.heatmap(passengers_matrix , annot = True , fmt = "d" , annot_kws={"size": 6})
```

```python

# 指定格子之间的间距linewidths = .5

ax = sns.heatmap(passengers_matrix , linewidths = .5)

```

```python

# 指定调色板
# YlGnBu (Yelllow-Green-Blue)
# YlOrRd (Yellow-Orange-Red) 
# RdYlBu (Red-Yellow-Blue) 

# Blues	单色系蓝色	简洁商务风
# coolwarm	冷暖对比色	正负值数据（如相关性矩阵）
# viridis	彩虹渐变（色盲友好）	通用数据可视化
# rocket	Seaborn 暖色系	热度、计数类数据

ax = sns.heatmap(passengers_matrix  , linewidths = .5 , cmap = "viridis")
```
