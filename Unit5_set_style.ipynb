```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def sinplot(filp = 1):
    x = np.linspace(0 , 14 , 100) # np.linspace函数可以生成一个指定范围内的等差数列，参数分别是起始值，结束值，元素个数。
    for i in range(1,7):
        plt.plot(x , np.sin(x + i * .5) * (7 - i) * filp) #sin函数在numpy库中，调用就要np.
```

```python
sinplot()
plt.show()
```

```python
sinplot(filp = 1)
sns.set()
plt.show()
```

```python
# whitegrid , 白色背景板

sns.set_style("whitegrid")

sinplot()

plt.show()
```

```python
sns.set_style('whitegrid') #设置背景样式

data = np.random.normal(size = (20 , 6)) + np.arange(6) / 2 #生成一个20行6列的正态分布数据，后面是为了增加数据的差异性

sns.boxplot(data = np.log(data))

plt.show()
```

```python
# dark , 去掉刻度线

sns.set_style("dark")

sinplot()

plt.show()
```

```python
# white , 去掉背景线条并且背景变成white

sns.set_style("white")

sinplot()

plt.show()
```

```python
# ticks ， 去掉背景线条，增加坐标轴上刻度尺

sns.set_style("ticks")

sinplot()

sns.despine() # 默认去掉右边与上方的线条

plt.show()
```

```python
# ticks ， 去掉背景线条，增加坐标轴上刻度尺

sns.set_style("ticks")

sinplot()

sns.despine(offset=30) #去掉坐标轴上方和右方的线条，offset参数可以设置线条和坐标轴的距离

plt.show()
```

```python
sns.set_style('whitegrid') #设置背景样式

data = np.random.normal(size = (20 , 6)) + np.arange(6) / 2 #随机生成一个20行6列的正态分布数据，后面是为了增加数据的差异性

sns.boxplot(data = np.log(data) )

sns.despine(left = True , bottom=True) # 去掉左边与下边的线条

plt.show()
```

```python
# 指定图像的风格，线条粗细，刻度大小等参数
sns.set_style("white")

# paper-0.8 , notebook-1.0, talk-1.3 ,  poster-1.6,从小到大,相较于标准notebook，字体，外框其余等比放大缩小 ; 
# font_scale参数可以设置字体的大小 ; 
# rc参数可以设置线条的粗细，参数是一个字典，key是要设置的参数，value是要设置的值
sns.set_context("notebook" , font_scale = 1.5 , rc = {"lines.linewidth" : 2.5}) 

fig = plt.figure(figsize = (8,6))

sns.despine()

sinplot()

plt.show()

```
