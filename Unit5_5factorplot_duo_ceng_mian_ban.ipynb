```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = pd.read_csv(r"D:\JupyterNotebook\PythonFile\Seaborn\set1.csv")

tips[2:15:2]

tips["tip"][1:4]
```

```python
# 分类

# kind are 'strip'(点图), 'swarm'(树状点图), 'box'(盒图), 'boxen', 'violin'(小提琴图)(split = True), 'bar'(柱状图), 'count'(单变量分析,统计数量), and 'point'

#  count : sns.catplot(data = tips ,  y = "day" , hue = "sex" , palette = {"Female" : "b" , "Male" : "y"} , kind = "count")
fig = plt.figure(figsize = (8 , 16))

# ax = [None] * 8

# for i in range(8):
#     ax[i] = fig.add_subplot(4,2,i + 1)


sns.catplot(data = tips ,x = "size" , y = "tip" , hue = "sex" , 
palette = {"Female" : "b" , "Male" : "y"} , 
kind = "point" , col = "time") # col = "",指定一个需要参考的值，或者说要分类探索


# paper-0.8 , notebook-1.0, talk-1.3 ,  poster-1.6,从小到大,相较于标准notebook，字体，外框其余等比放大缩小 ; 
# font_scale参数可以设置字体的大小 ; 
# rc参数可以设置线条的粗细，参数是一个字典，key是要设置的参数，value是要设置的值
sns.set_context("notebook" , font_scale = 1 , rc = {"lines.linewidth" : 1}) 


sns.despine(offset=10) #去掉坐标轴上方和右方的线条，offset参数可以设置线条和坐标轴的距离

# ax1 = sns.~~~
# ax1.tick_params(axis = "x" , rotation = 30)
```

```python
# 所有属性之间的关系，pairplot

# data	必需，Pandas DataFrame 数据	-
# hue	按分类变量分组着色（最常用）	列名，如 'species'
# vars	仅绘制指定的数值列	列表，如 ['sepal_length', 'petal_width']
# kind	非对角线图类型	'scatter'(默认), 'reg'(带回归线), 'kde'(核密度图), 'hist'（直方图） ,  
# diag_kind	对角线图类型	'auto', 'hist', 'kde'
# palette	配色方案	'husl', 'viridis', 'Set2'
# height / aspect	子图高度 / 宽高比	数字，如 height=2
# corner	只画左下角三角（节省空间）	True / False
# plot_kws	自定义非对角线样式	{'alpha': 0.5, 's': 30} (透明度 / 点大小)

sns.pairplot(data = tips , kind = "reg")
```

```python
# 单变量分析

# 数据量大观测深浅观察变量之间的关系，重复度越高越深
sns.set_style("ticks")

sns.jointplot(x = "total_bill" , y = "tip" , data = tips , kind = "hex" , color = "green") # kind = "hex"表示画出六边形的散点图

sns.set_context("notebook" , font_scale = 1 , rc = {"lines_linewidth" : .5})

plt.show()
```

```python

#  displot单变量分析，频率直方图

total_bill = tips["total_bill"]

# 设置样式
sns.set_style("whitegrid")
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})

sns.distplot(total_bill , kde = True  , bins = 20 , color = "r")

# 移除多余边框
sns.despine()
plt.show()
```
