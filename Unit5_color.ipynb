```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# color_palette() --- 调色板 ， 能传入matplotlib中所有颜色

currrent_palette = sns.color_palette()
sns.palplot(currrent_palette)
```

```python
# 渐变色，使用hls色系，30种颜色

sns.palplot(sns.color_palette("hls" , 15))
```

```python
sns.set_style("ticks")

data = np.random.normal(size = (20 , 12)) + np.arange(12) / 2

sns.boxplot(data = np.log(data) , palette = sns.color_palette("hls" , 12))


sns.despine()

sns.set_context("paper" , font_scale = 1.5 , rc = {"lines.linewidth" : 2.5})

plt.show()
```

```python
# hls色系 --- 色相，亮度，饱和度  l-亮度 ， s-饱和度
sns.palplot(sns.hls_palette(10 , l = .5 , s = .8))
```

```python
sns.set_style("ticks")

data = np.random.normal(size = (20 , 12)) + np.arange(12) / 2

sns.boxplot(data = np.log(data) , palette = sns.hls_palette(12 , l = .8 , s = .6))


sns.despine()

sns.set_context("paper" , font_scale = 1.5 , rc = {"lines.linewidth" : 2.5})

plt.show()
```

```python
# 颜色对

sns.palplot(sns.color_palette("Paired" , 12))
```

```python
# 连续画板

sns.palplot(sns.color_palette("Blues" , 10))

sns.palplot(sns.color_palette("BuGn_r" , 10))# _r表示反转
```

```python
x , y = np.random.multivariate_normal([2 , 0] , [[1 , .9] , [.9 , 1]] , size = 300).T

pal = sns.color_palette("Blues_r" , as_cmap = True)

sns.kdeplot(x = x , y = y , cmap = pal , levels = 10);

plt.show()
```
