```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style("whitegrid")

x = np.random.normal(size =1000)

fig = plt.figure(figsize = (8,6))

sns.set_context("notebook" , font_scale = 1.5 , rc = {"lines_linewidth" : 2.5})

sns.despine()

sns.distplot(x , kde = True ) # kde = True表示画出平滑曲线和直方图 ， False表示只画出直方图

plt.show()
```

```python
sns.distplot(x , bins = 50 , kde = True , color = "red")

plt.show()
```

```python
mean = [0,1]
cov = [[1 , .5] , [.5 , 1]]

data = np.random.multivariate_normal(mean , cov , size = 1000)

df = pd.DataFrame(data , columns = ["x" , "y"])

df
```

```python
# 观测两变量之间的关系，散点图 matplotlib中plt.scatter
# sns.jointplot(x = "x" , y = "y" , data = df) # 画出散点图和核密度图
sns.set_style("ticks")

sns.jointplot(x = "x" , y = "y" , data = df ) # kind = "hex"表示画出六边形的散点图

sns.set_context("notebook" , font_scale = 1 , rc = {"lines_linewidth" : .5})

plt.show()

```

```python
# 数据量大观测深浅观察变量之间的关系，重复度越高越深
sns.set_style("ticks")

sns.jointplot(x = "x" , y = "y" , data = df , kind = "hex" , color = "green") # kind = "hex"表示画出六边形的散点图

sns.set_context("notebook" , font_scale = 1 , rc = {"lines_linewidth" : .5})

plt.show()

```
